```python
import sklearn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_predict, train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
from tpot import TPOTRegressor
```

```python
df = pd.read_csv('housing.csv')
X = df.iloc[:, 1:]
y = df['price']
```

```python
def yesno_to_binary(df, cols):
    for c in cols:
        df[c] = df[c].replace({
            'yes': 1,
            'no': 0
        })
    return df

X = yesno_to_binary(X, [
    'mainroad',
    'guestroom',
    'basement',
    'hotwaterheating',
    'airconditioning',
    'prefarea'
])

X.head()
```

```python
X = pd.concat([X, pd.get_dummies(X['furnishingstatus'], drop_first=True)], axis=1)
```

```python
X = X.drop('furnishingstatus', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
tpot = TPOTRegressor(generations=10, population_size=50, verbosity=2, random_state=42)
tpot.fit(X_train, y_train)
```

```python
y_pred = tpot.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```
